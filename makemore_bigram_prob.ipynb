```python
from numpy import dtype
from sympy.codegen.ast import int32

words = open('names.txt', 'r')
wordlist = list(words.read().splitlines())
wordlist[:3]
```

```python
s = ['.'] + sorted(set(i for w in wordlist for i in w))
s
stoi = {}
itos = {}
for i, j in enumerate(s):
    stoi[j] = i
    itos[i] = j
print(stoi['a'])
itos[1]
```

```python
b = {}
for w in wordlist:
    word = '.' + w + '.'
    for c1, c2 in zip(word, word[1:]):
        b[(c1, c2)] = b.get((c1, c2), 0) + 1
sorted(b.items(), key=lambda x: -x[1])
```

```python
import torch
```

```python
N = torch.zeros([27, 27], dtype=torch.int32)
for word in wordlist:
    w = '.' + word + '.'
    for c1, c2 in zip(w, w[1:]):
        N[stoi[c1]][stoi[c2]] += 1
N
```

```python
P = N.float()
P /= P.sum(1, keepdims=True)
```

```python
g = torch.Generator().manual_seed(2147483647)

for i in range(5):
    ix = 0
    out = []
    while True:
        ix = torch.multinomial(P[ix], 1, True, generator=g).item()
        if ix == 0:
            break
        out.append(itos[ix])
    print(*out, sep='')
```

```python
log_likelihood = 0.
n = 0
for w in wordlist:
    word = '.' + w + '.'
    for c1, c2 in zip(word, word[1:]):
        ix1, ix2 = stoi[c1], stoi[c2]
        prob = P[ix1][ix2]
        log_likelihood += prob.log()
        n += 1
print(f"{log_likelihood=}")
nll = -log_likelihood
print(f"{nll=}")
print(f"avg: {nll/n}")
```
